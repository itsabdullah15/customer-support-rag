Geting All data from this WebPage: https://www.angelone.in/support

```python
import requests
from bs4 import BeautifulSoup

url = "https://www.angelone.in/support"
headers = {"User-Agent": "Mozilla/5.0"}

response = requests.get(url, headers=headers)
html_content = response.text

soup = BeautifulSoup(html_content, "html.parser")
quick_ten_links = soup.find_all("div", class_="carousel-cell")

data = []

for enum, div in enumerate(quick_ten_links, 1):
    if div.a:
        text = div.a.text.strip()
        link = div.a.get("href")

        # Make full URL if relative
        if link.startswith("/"):
            link = "https://www.angelone.in" + link

        try:
            response1 = requests.get(link, headers=headers)
            response1.raise_for_status()
            soup1 = BeautifulSoup(response1.text, "html.parser")
            detail_text = soup1.get_text(strip=True)

            # Add to list as dictionary
            data.append({
                "Text": text,
                "Link": link,
                "Details": detail_text
            })

            print(enum, "✓ Scraped:", text)
        except Exception as e:
            print(enum, "✗ Failed:", text, "| Error:", e)

# Optional: Save to JSON or print
import json

with open("angelone_quick_10_links_support_data.json", "w", encoding="utf-8") as f:
    json.dump(data, f, indent=2, ensure_ascii=False)
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://www.angelone.in/support"
headers = {"User-Agent": "Mozilla/5.0"}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")

# Find all the divs with class 'grid'
grid_divs = soup.find_all("div", class_="grid")

data = []

for div in grid_divs:
    # Get the first <a> tag with href
    link_tag = div.find("a", href=True)
    if not link_tag:
        continue

    link = link_tag["href"]

    # Make sure it's a full URL
    if link.startswith("/"):
        link = "https://www.angelone.in" + link

    # Find the h2 and p tags inside the div
    heading_tag = div.find("h2")
    description_tag = div.find("p")

    heading = heading_tag.get_text(strip=True) if heading_tag else ""
    description = description_tag.get_text(strip=True) if description_tag else ""

    # Now get the inner page text
    try:
        page_response = requests.get(link, headers=headers)
        page_response.raise_for_status()
        page_soup = BeautifulSoup(page_response.text, "html.parser")
        page_text = page_soup.get_text(separator="\n", strip=True)
    except Exception as e:
        page_text = f"Error fetching page: {e}"

    data.append({
        "Text": heading,
        "Description": description,
        "Link": link,
        "Details": page_text
    })

# Print a summary
for item in data:
    print(f"Title: {item['Text']}")
    print(f"Description: {item['Description']}")
    print(f"Link: {item['Link']}")
    print(f"Details (excerpt): {item['Details'][:300]}...\n")  # Preview only


import json

with open("angelone_support_full_data.json", "w", encoding="utf-8") as f:
    json.dump(data, f, indent=2, ensure_ascii=False)
```

```python
import os
import json
from langchain.document_loaders import PyPDFLoader

folder_path = "Insurance PDFs"
all_documents = {}

# Loop through all PDFs in the folder
for filename in os.listdir(folder_path):
    if filename.endswith(".pdf"):
        pdf_path = os.path.join(folder_path, filename)
        print(f"Loading: {filename}")

        loader = PyPDFLoader(pdf_path)
        documents = loader.load()

        pages = []
        for doc in documents:
            pages.append({
                "page_number": doc.metadata.get("page", None),
                "content": f"[{filename}] {doc.page_content.strip()}"
            })

        all_documents[filename] = pages

# Save to JSON file
with open("insurance_pdfs_text1.json", "w", encoding="utf-8") as f:
    json.dump(all_documents, f, indent=2, ensure_ascii=False)

print("\n✅ PDF contents saved to 'insurance_pdfs_text.json'")
import os
import json
from langchain.document_loaders import PyPDFLoader

folder_path = "Insurance PDFs"
all_pages = []

# Loop through all PDFs in the folder
for filename in os.listdir(folder_path):
    if filename.endswith(".pdf"):
        pdf_path = os.path.join(folder_path, filename)
        print(f"Loading: {filename}")

        loader = PyPDFLoader(pdf_path)
        documents = loader.load()

        for doc in documents:
            all_pages.append({
                "Text": filename,
                "page_number": doc.metadata.get("page", None),
                "Details": doc.page_content.strip()
            })

# Save to JSON file
with open("insurance_pdfs_text.json", "w", encoding="utf-8") as f:
    json.dump(all_pages, f, indent=2, ensure_ascii=False)

print("\n✅ PDF contents saved to 'insurance_pdfs_text.json'")
```
